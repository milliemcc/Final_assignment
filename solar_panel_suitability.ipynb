## Introduction


The Climate Emergency is a severe challenge facing the globe (Ripple et al., 2020), and action must be taken to reduce greenhouse gas emissions to limit global temperature rise to the agreed level of 1.5oC above pre-industrial levels (United Nations, 2015). As part of this effort, the Scottish Government aim to reach Net Zero by 2045 under the Scottish Climate Change Act of 2009 (Scottish Government, 2009). Under this goal, Scotland must decarbonise its energy supply by shifting to a predominantly renewable energy mix, however, land conflicts and requirements limit this potential. This study will examine the suitability of solar farm development in Fife, Scotland, while accounting for potential tension over land use with productive agricultural land.

Despite recent progress to increase the proportion of renewable energy, 11.4% of Scotland’s energy supply was derived from fossil fuels in 2024. Solar farms using photovoltaic energy are widely acknowledged to play a key role in this transition by sourcing a sustainable and renewable form of energy (Maka and Alabid, 2022) and both ground and roof mounted photovoltaic cells can contribute to this. Ground mounted solar farms have previously been installed across the UK and are acknowledged as one of the most affordable forms of renewable energy, encouraging the UK government to aim to quintuple solar energy potential by 2035 (HM Government, 2023). Their low-cost makes solar farms a highly attractive energy source, creating a high demand to build more ground mounted solar farms.

While there is a high demand for more solar power, this is restricted by a range of social, environmental, economic and technical factors that must be considered before constructing a solar farm (Palmer et al., 2019). Specifically, land with high biodiversity and ecological value must be avoided as despite the positive impact of solar farms on emissions, this will cause more damage, making efforts counter-productive (Scottish Government, 2013). This is especially true for Scotland’s carbon-rich peatlands. Additionally, the impact on the local community is essential to consider as their perception determines the success of a project (Palmer et al., 2019).

Within the UK context, one of the biggest concerns surrounding renewable energy is their tension with agricultural land (Rankl, 2024). The future of Scotland’s agriculture is a growing concern within the context of global instability and climate change (Department for Environment, Food and Rural Affairs, 2024) and efforts must be made to protect the livelihoods of farmers. However, there is a large overlap between the land best suited for agriculture, and the land best suited for the development of a solar farm. Scotland faces the challenge of balancing a self-sufficient food network with the need to expand renewable energy sources to meet current policy goals.

This research aims to understand the suitability of Fife, Scotland, for expanding the solar energy network while simultaneously exploring the extent of its conflict with agriculture. This research is guided by the following questions:

•	What is the suitability of developing a solar farm in Fife?

•	To what extent does agriculture limit solar farm expansion in Fife?

### Solar farm MCE

To address the above questions, two multi-criteria Evaluations (MCEs) were created, as well as subsequent conflict maps at varying levels of suitability. The MCEs were for the suitability of the landscape for solar energy and agriculture respectively, allowing for comparison and analysis between the two activities. QGIS was utilised to process data and produce maps, using a range of its available tools to prepare layers to reflect the required criteria for each MCE. All layers were clipped to the outline of Fife. The following section will lay out the data and criteria for each MCE.

To create an MCE for solar farm development, six criteria were used. After consulting the literature and government guidance above, the following criteria were established: proximity to roads, distance from communities, sunlight (measured using irradiance), appropriate land use, conservation areas, and an appropriate slope. While there is undoubtably further criteria that would aid understanding and accuracy, data limitations and scope have prevented their inclusion in the MCE.

Roads were selected as a factor since access to the site is essential for the construction process of solar farms. The greatest environmental impact takes place during the construction of a solar farm (Solar Energy UK, 2024) and access roads must be built to reach the site. Therefore, to reduce the environmental disruption and cost, it is recommended that a solar farm site is located close to a road, as found in Figure 2.


```python
roads = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/NormalisedRoads.tif')
plt.figtext(0,0, "Figure 2: raster for the factor critierion for proximity to roads. White shows greater proximity to the road, and conseqently greater suitability (OS, 2024)")
show(roads, title='Figure 2: Roads factor', cmap="gist_grey")
```

The aesthetic, and glint and glare impact (Ho et al., 2011) of solar farms are of large concern to neighbouring settlements, and their impacts should be mitigated. It was assumed this would be most reduced if a solar farm is located further from a settlement. Land use data was used to identify settlement areas (see Appendix 1), and the distance furthest from here was identified as suitable, highlighted in white in Figure 3

```python
visibility = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/NormalVis.tif')
plt.figtext(0,0, "Figure 3: Raster for the factor criterion showing proximity of a solar farm development to a community. This is a proxy for the visibility and glint impact of the development (Copernicus Land, 2020)")
show(visibility, title='Figure 3: Visibility factor', cmap='gist_grey')
```

The long-term yearly average of direct normal irradiation was used to represent the amount of sunlight an area receives. Figure 4 highlights the areas of Fife that receive the most sunlight and are most suitable for a higher solar energy yield.  

```python
sunlight = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/NormalSun.tif')
plt.figtext(0,0, "Figure 4: Raster for the factor criterion of yearly direct irradiation across Fife with white showing the most irradiation, subsequent suitability (Global Solar Atlas, 2023)")
show(sunlight, title='Figure 4: Sunlight factor', cmap='gist_grey')
```

Land use is a constraint criterion whereby the land uses not suitable were assigned a value of zero. The Corine Land Cover dataset (Copernicus Land, 2018) was used for this criterion, consisting of 44 classifications (Appendix 1), and of these, classes depicting urban uses, forest, marshes and peat bogs were deemed unsuitable. Their distribution across Fife is found in Figure 5.

```python
landuse = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/Landuse_criteria.tif')
plt.figtext(0,0, "Figure 5: Raster showing the land use constrain criterion, whereby the areas in black are excluded from the MCE due to unsuitability. These show areas of urban use, forest, marsh and peat (Copernicus Land, 2020)")
show(landuse, title='Figure 5: Land use constraint', cmap='gist_grey')
```

Land under conservation was excluded from consideration for solar farms due to its ecological significance that outweighs the environmental benefit of a renewable energy source in that location. A shapefile specific to conservation areas in Fife was used and these areas are outlined in Figure 6.

```python
conservation = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/Conservation_bool.tif')
plt.figtext(0,0, "Figure 6: Constraint raster showing conservation areas in black, which are deemed unsuitable for development (Scottish Spatial Data, 2024)")
show(conservation, title='Figure 6: Conservation factor', cmap='gist_grey')
```

A slope constraint was included due to limitations in the physical construction and planning policy for solar farms (Palmer et al., 2019). Literature places the slope constraint between 2% and 11% (Ibid.). In this instance, the slope was limited to any area less than a 5% gradient, as found in figure 7.

To calculate the MCE, assuming they share similar importance, equal weightings were assigned to all factors, resulting in the following calculation:

(1/3 * Roads + 1/3 * proximity to community + 1/3 * sunlight) * land use * conservation area * slope


```python
slope = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Solar/AppropriateSlope.tif')
plt.figtext(0,0, "Figure 7: Raster showing slope constrain where dark areas have a sloe over 5% so are unsuitable for development. Taken from Lab 3: Scotland_DEM")
show(slope, title='Figure 7: Slope constraint', cmap='gist_grey')
```

### MCE for agriculture

To identify the best land for agriculture, four criteria were used: proximity to rivers, agricultural grade, slope constraint and flatter slope factor. A similar process to the above MCE was carried out to produce this.  

Proximity to rivers was selected due to the importance of water and irrigation to agricultural productivity (Figure 8). The closer an area of agricultural land is to a river; it will likely have improved soil quality and therefore bw more appropriate for agriculture.


```python
rivers = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Agriculture/Rivers crit.tif')
plt.figtext(0,0, "Figure 8: Raster showing rivers factor criterion whereby white is closest to the river and so most suitable (OS, 2024)")
show(rivers, title='Figure 8: Rivers factor', cmap="gist_grey")
```

Land productivity was deemed a highly significant criterion to identify the most suitable agricultural land. A dataset showing land capability for agriculture combines soil quality with land type to produce a classification rank for the land most productive for agriculture (Soil Survey of Scotland, 1984-87). Here, Classes 1 to 4 were determined suitable for agriculture (see Appendix 2). Class 5 and above were deemed to have a limited capability for producing crops, therefore were excluded from the MCE, as found in Figure 9.

```python
agri = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Agriculture/suitable_agri_criteria.tif')
plt.figtext(0,0, "Figure 9: Raster showing agriculture classification constraint. Areas in black have an land capability class of 5 and above so is unsuitable for agriculture (Scottish Spatial Data, 2016)")
show(agri, title="Figure 9: High agricultural grade constraint", cmap="gist_grey")
```

A slope factor and constraint criteria were also included. Flatter land is most suited to agriculture and slopes greater than 7% are deemed unsuitable for farming difficult (UK Government, 2024) so are excluded in a constraint (Figure 10). Additionally, the flatter the land, the more suitable it is for agriculture; therefore, a factor criterion was used whereby the shallower the slope, the more appropriate it is (Figure 11).

```python
slope_bool = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Agriculture/agri_slope_bool.tif')
plt.figtext(0,0, "Figure 10: Raster showing slope constraint. Areas in black have a slope greaer than 7% so are unsuitable for agriculture. From Lab 3: Scotland_DEM")
show(slope_bool, title='Figure 10: Slope constraint', cmap='gist_grey')
```

```python
agri_slope = rio.open('/content/drive/MyDrive/Colab Notebooks/final_assessment/Data/Agriculture/Fife_normalised_slope.tif')
plt.figtext(0,0, "Figure 11: Raster showing slope factor for slope. White represents a flat slope so is more suitable for agriculture. From Lab 3: Scotland_DEM")
show(agri_slope, title='Figure 11: Slope factor', cmap='gist_grey')
```

To calculate this MCE, equal weights were assigned to both factors as were deemed equally important to the suitability of land for agriculture, creating the following equation:

(0.5 * rivers + 0.5 * slope factor) * land capability * slope constraint




To visually identify areas where both MCEs find suitable, conflict maps were produced at three levels of suitability: 70%, 80% and 90%. This visually displays areas of potential conflict.

## Discussion

These results confirm the above concerns over the potential for conflict between the demand for solar farms and the need to protect food security. While the extent of the area where both land uses are very suitable (90%) is more limited, there is a large area of land where both land uses are 70% suitable. The similarities in land requirements, including shallow slopes and non-urban areas offer a reason for this overlap and why it is predominantly in East Fife, which is a typically flat and rural area, as opposed to West Fife with a high concentration of large urban areas, such as Dunfermline.

As solar farm development has an on average slightly lower suitability than agriculture, it appears more challenging to develop a solar farm on land that is not of high suitability for agriculture using these criteria, confirming concerns over balancing these two important land uses. As conflict is likely in Fife, steps should be taken to mediate tension to allow for coexistence. While it may appear that the two land uses are broadly incompatible, there is evidence to they are not mutually exclusive, and rather land can be “borrowed” for solar farms and returned to agricultural land use after decommissioning (Solar Energy UK, 2022). This may benefit soil from allowing it to recover from intensive farming practices, improving longevity. Livestock grazing is also possible without disrupting the functioning of a solar farm (BRE, 2014), offering solutions to cope with possible future tensions.

The differences in suitability between the MCEs can be attributed to differences in the criteria used. Solar farms have more criteria, including factors with varied suitability, whereas that of agriculture is more reduced. The factors included in agriculture are also more indicative of suitability. Additionally, there are limits to the agricultural grade criteria which is likely affecting the outcome. Grades 1 and 4 are treated equally despite the large difference of land productivity. To improve the outcome of the conflict map, comparing the results using different grades, especially solely Grades 1 and 2 to identify the most important land for crop production where compromise between the two land uses is less likely.  

A further limitation is the range of criteria in each MCE. There are likely other factors relevant to establishing suitability which have been excluded from the study. Future research should focus on identifying additional values. Furthermore, each factor was weighted equally in the MCE calculation as were assumed to hold equal important. Holding a consultation with local farmers and Fife council will improve understanding of criteria, allowing different weightings to be negotiated. The visibility factor used distance from settlements as a proxy for the aesthetic and glint and glare impacts. However, combining a range of criteria, as well as local opinion may be a better guide to estimate this impact. To mitigate the significance of these limitations, on-site analysis and consultation is essential to understand the wider context of the area.

Finally, this study did not specify the size of the solar farm to be developed but rather the general suitability of the land for solar energy. As different sizes of farms require different planning permissions and criteria (Rankle, 2024), future research should examine the suitability of these different sizes, and their resulting conflict with agriculture.



## Conclusion

This study shows that Fife is a fairly suitable region for solar farm development, identifying East Fife as the most suitable. Throughout Fife, land was found to be very suitable for agriculture, creating overlaps with solar energy, inferring future land use tensions. Land suitable for solar farm development covers a smaller area than agriculture, and in almost every area where solar farms are suitable, so too is agriculture. Therefore, an effort must be made to limit the severity of this conflict to balance the need to protect high-quality agricultural land with expanding Scotland’s renewable energy supply. Developers must work closely with farmers to find solutions for each land use to coexist in Fife.
